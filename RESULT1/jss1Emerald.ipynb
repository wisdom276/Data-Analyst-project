```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
JS1_Eme = pd.read_csv('JSS1 Emerald.csv')
```

```python
JS1_Eme
```

```python
del JS1_Eme ['Unnamed: 0']
```

```python
JS1_Eme
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="Student Name", y="Average (%)", data=JS1_Eme, palette="viridis")
plt.xticks(rotation=45, ha="right")
plt.title("Top Performers' Average Scores", fontsize=16)
plt.xlabel("Student Name", fontsize=12)
plt.ylabel("Average (%)", fontsize=12)
plt.tight_layout()
plt.show()
```

Pie Chart for Grade Distribution



```python
grade_counts = JS1_Eme['Grade'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(grade_counts, labels=grade_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Grade Distribution")
plt.show()
```

Line Chart for Average Scores vs. Position

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x="Position", y="Average (%)", data= JS1_Eme, marker="o")
plt.title("Average Scores by Position", fontsize=16)
plt.xlabel("Position", fontsize=12)
plt.ylabel("Average (%)", fontsize=12)
plt.grid()
plt.show()
```

```python
Observations:
Top Performers excelled in Civic Education, Ibibio, and Home Economics.
Struggling students frequently failed Mathematics, Computer Studies, and History.
A strategic focus on STEM subjects and foundational literacy is crucial for improvement.
```
