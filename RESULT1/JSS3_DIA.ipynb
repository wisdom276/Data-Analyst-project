```python
grade_counts = JS3_DIA['Grade'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(grade_counts, labels=grade_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Grade Distribution")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x="Position", y="Average (%)", data= JS3_DIA, marker="o")
plt.title("Average Scores by Position", fontsize=16)
plt.xlabel("Position", fontsize=12)
plt.ylabel("Average (%)", fontsize=12)
plt.grid()
plt.show()
```
