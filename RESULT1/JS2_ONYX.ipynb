```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
JS2_ONYX = pd.read_csv('JS2_ONYX.csv')
```

```python
JS2_ONYX
```

```python
# How to Print number of students with B grade
b_grade_students = JS2_ONYX[JS2_ONYX['Grade'] == 'B']
```

```python
grade_summary = JS2_ONYX['Grade'].value_counts()
```

```python
grade_summary
```

```python
grade_counts = JS2_ONYX['Grade'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(grade_counts, labels=grade_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Grade Distribution")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x="Position", y="Average (%)", data= JS2_ONYX , marker="o")
plt.title("Average Scores by Position", fontsize=16)
plt.xlabel("Position", fontsize=12)
plt.ylabel("Average (%)", fontsize=12)
plt.grid()
plt.show()
```
