```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
JSS2_DIA = pd.read_csv('JSS2_DIA.csv')
```

```python
JSS2_DIA 
```

```python
# How to Print number of students with B grade
b_grade_students = JSS2_DIA[JSS2_DIA['Grade'] == 'B']
```

```python
b_grade_students 
```

```python
grade_summary = JSS2_DIA['Grade'].value_counts()
```

```python
grade_summary
```

```python
grade_counts = JSS2_DIA['Grade'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(grade_counts, labels=grade_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Grade Distribution")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x="Position", y="Average (%)", data= JSS2_DIA, marker="o")
plt.title("Average Scores by Position", fontsize=16)
plt.xlabel("Position", fontsize=12)
plt.ylabel("Average (%)", fontsize=12)
plt.grid()
plt.show()
```
