```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
JS2_SAPP = pd.read_csv('JS2_SAPP.csv')
```

```python
JS2_SAPP
```

```python
# How to Print number of students with B grade
b_grade_students = JS2_SAPP[JS2_SAPP['Grade'] == 'B']
```

```python
b_grade_students
```

```python
grade_summary = JS2_SAPP['Grade'].value_counts()
```

```python
# Display the summary
print(grade_summary)
```

```python
grade_counts = JS2_SAPP['Grade'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(grade_counts, labels=grade_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Grade Distribution")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x="Position", y="Average (%)", data= JS2_SAPP, marker="o")
plt.title("Average Scores by Position", fontsize=16)
plt.xlabel("Position", fontsize=12)
plt.ylabel("Average (%)", fontsize=12)
plt.grid()
plt.show()
```
