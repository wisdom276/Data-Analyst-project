```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
JS1_DIA = pd.read_csv('JS1 DIA.csv')
```

```python
JS1_DIA
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="Student Name", y="Average (%)", data=JS1_DIA, palette="viridis")
plt.xticks(rotation=45, ha="right")
plt.title("Top Performers' Average Scores", fontsize=16)
plt.xlabel("Student Name", fontsize=12)
plt.ylabel("Average (%)", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
grade_summary = JS1_DIA['Grade'].value_counts()
```

```python
grade_summary
```

```python
# Piechart visualization
```

```python
grade_counts = JS1_DIA['Grade'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(grade_counts, labels=grade_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Grade Distribution")
plt.show()
```

```python
#Line chart
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x="Position", y="Average (%)", data= JS1_DIA, marker="o")
plt.title("Average Scores by Position", fontsize=16)
plt.xlabel("Position", fontsize=12)
plt.ylabel("Average (%)", fontsize=12)
plt.grid()
plt.show()
```
