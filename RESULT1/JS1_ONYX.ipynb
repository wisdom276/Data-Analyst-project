```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
JS1_ONYX = pd.read_csv('JS1_ONYX.csv')
```

```python
JS1_ONYX
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="Student Name", y="Average (%)", data=JS1_ONYX, palette="viridis")
plt.xticks(rotation=45, ha="right")
plt.title("Top Performers' Average Scores", fontsize=16)
plt.xlabel("Student Name", fontsize=12)
plt.ylabel("Average (%)", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
grade_counts = JS1_ONYX['Grade'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(grade_counts, labels=grade_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Grade Distribution")
plt.show()
```

OBSERVATIONS

```python
Observations
Top Performers:

Students ranked 1st to 5th demonstrated consistent performance in Coding, Home Economics, and Ibibio, with averages above 80%.
The top students still have room for improvement in STEM subjects like Mathematics and Basic Technology.
Low Performers:

Students ranked 13th–20th performed poorly across multiple subjects, especially Mathematics, English Language, and Computer Studies.
Persistent failures in STEM and language-related subjects indicate systemic weaknesses.
Subject Strengths:

Strong performances were observed in Home Economics, Cultural Arts, and Coding, reflecting effective teaching or student interest in these areas.
Subject Weaknesses:

Mathematics, English Language, and Basic Technology show widespread underperformance, even among top-ranking students.
Technical subjects like Computer Studies and Practical Agriculture need significant attention.
Psychomotor and Social Skills:

Students generally performed well in drama, sports, and obedience.
Attentiveness and problem-solving need improvement in the lower-ranked students.
This analysis highlights both academic strengths and weaknesses, providing a foundation for targeted interventions.
```
