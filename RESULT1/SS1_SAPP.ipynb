```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
SS1_SAPP = pd.read_csv('SS!_SAPP.csv')
```

```python
SS1_SAPP
```

```python
# How to Print number of students with B grade
b_grade_students = SS1_SAPP[SS1_SAPP['Grade'] == 'B']
```

```python
b_grade_students
```

```python
grade_summary = SS1_SAPP['Grade'].value_counts()
```

```python
grade_summary
```

```python
grade_counts = SS1_SAPP['Grade'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(grade_counts, labels=grade_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Grade Distribution")
plt.show()
```
