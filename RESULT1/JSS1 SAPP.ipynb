```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
JS1_SAPP = pd.read_csv('JSS1_Sapp_Res_ana.csv')
```

```python
JS1_SAPP
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="Student Name", y="Average (%)", data= JS1_SAPP, palette="viridis")
plt.xticks(rotation=45, ha="right")
plt.title("Top Performers' Average Scores", fontsize=16)
plt.xlabel("Student Name", fontsize=12)
plt.ylabel("Average (%)", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
grade_counts = JS1_SAPP['Grade'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(grade_counts, labels=grade_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Grade Distribution")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x="Position", y="Average (%)", data= JS1_SAPP, marker="o")
plt.title("Average Scores by Position", fontsize=16)
plt.xlabel("Position", fontsize=12)
plt.ylabel("Average (%)", fontsize=12)
plt.grid()
plt.show()
```

```python
Observations
Top Performers:

Students ranked 1st–5th excelled in subjects like French, Home Economics, and Civic Education with averages above 80%.
Consistent good grades reflect strong foundational knowledge and effective teaching.
Low Performers:

Students ranked 16th–20th struggled significantly, particularly in Mathematics, English Language, and Computer Studies.
These students require targeted interventions.
Subject Trends:

Strengths:
High performances in Home Economics, French, and Civic Education indicate interest and good teaching methods.
Weaknesses:
Low scores in Mathematics, Basic Technology, and Computer Studies highlight systemic gaps in understanding and engagement.
```
