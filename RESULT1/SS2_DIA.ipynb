```python
plt.figure(figsize=(10, 6))
sns.lineplot(x="Position", y="Average (%)", data= SS2_DIA, marker="o")
plt.title("Average Scores by Position", fontsize=16)
plt.xlabel("Position", fontsize=12)
plt.ylabel("Average (%)", fontsize=12)
plt.grid()
plt.show()
```
